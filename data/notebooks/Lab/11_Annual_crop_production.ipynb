# Superficies y producciones anuales de cultivos

[Datos 2020](https://www.mapa.gob.es/es/estadistica/temas/estadisticas-agrarias/agricultura/superficies-producciones-anuales-cultivos/)

```python
import pandas as pd
```

## Cereales

```python
df_1 = pd.read_excel('../../datasets/raw/crops/cereal_2000-2010.xlsx', skiprows=4)
df_2 = pd.read_excel('../../datasets/raw/crops/cereal_2008-2018.xlsx', skiprows=4)

df_1 = df_1[['Años', 'Producción (miles de toneladas)']]
df_2 = df_2[['Años', 'Producción (miles de toneladas)']]

df = pd.concat([df_1, df_2[df_2['Años'] > 2010]])

df.rename(columns={"Años": "year", 'Producción (miles de toneladas)': "value"}, inplace=True)

df['dataset'] = 'Serie histórica de producción agrícola española'
df['indicator'] = 'Producción de cereales grano'
df['unit'] = 'kt'

df = df[['dataset', 'indicator', 'value', 'year', 'unit']]
df_cer = df.copy()
df_cer
```

## Olivar

```python
df_1 = pd.read_excel('../../datasets/raw/crops/olivar_2000-2010.xlsx', skiprows=6)
df_2 = pd.read_excel('../../datasets/raw/crops/olivar_2008-2018.xlsx', skiprows=6)

df_1 = df_1[['Años', 'Producción total de aceituna (miles de toneladas)']]
df_2 = df_2[['Años', 'Producción total de aceituna (miles de toneladas)']]

df = pd.concat([df_1, df_2[df_2['Años'] > 2010]])

df.rename(columns={"Años": "year", 'Producción total de aceituna (miles de toneladas)': "value"}, inplace=True)

df['dataset'] = 'Serie histórica de producción agrícola española'
df['indicator'] = 'Producción de aceituna'
df['unit'] = 'kt'

df = df[['dataset', 'indicator', 'value', 'year', 'unit']]
df_oliv = df.copy()
df_oliv
```

## Viñedo

```python
df_1 = pd.read_excel('../../datasets/raw/crops/viñedo_2000-2010.xlsx', skiprows=26).iloc[:-1]
df_2 = pd.read_excel('../../datasets/raw/crops/viñedo_2008-2018.xlsx', skiprows=26).iloc[:-1]

df_1 = df_1[['Años', 'Unnamed: 3']]
df_2 = df_2[['Años', 'Unnamed: 3']]

df = pd.concat([df_1, df_2[df_2['Años'] > 2010]])

df.rename(columns={"Años": "year", 'Unnamed: 3': "value"}, inplace=True)

df['dataset'] = 'Serie histórica de producción agrícola española'
df['indicator'] = 'Producción de uva'
df['unit'] = 'kt'

df = df[['dataset', 'indicator', 'value', 'year', 'unit']]
df_uva = df.copy()
df_uva
```

**Save tables**

```python
df = pd.concat([df_cer, df_oliv, df_uva]).reset_index(drop=True)
df.to_csv(f'../../datasets/processed/serie_historica_produccion_agricola_española.csv', index = False)
```
